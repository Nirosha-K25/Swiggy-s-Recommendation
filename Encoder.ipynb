```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
data_encoded = encoder.fit_transform(data[['city', 'cuisine']])
one_hot_df = pd.DataFrame(data_encoded, columns=encoder.get_feature_names_out(['city', 'cuisine']))
```

```python
encoded_data = pd.concat([data.loc[:,['cost','rating']], one_hot_df], axis=1)
encoded_data.head()
     
```

```python
encoded_data.shape
encoded_data.to_csv(r'encoded_data.csv', index=False)
encoded_data = pd.read_csv(r'encoded_data.csv')
encoded_data.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Fill NaN values with 0 before computing similarity
# encoded_data_filled = encoded_data.fillna(0)

# Function to get top N similar restaurants for a given index
def get_similar_restaurants(idx, top_n=5):
    # Compute cosine similarity between the target and all others (1 row vs all)
    target_vector = encoded_data.iloc[[idx]]
    sim_scores = cosine_similarity(target_vector, encoded_data)[0]
    # Pair index with score and sort
    sim_scores = list(enumerate(sim_scores))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Exclude the restaurant itself (idx)
    sim_indices = [i for i, score in sim_scores[1:top_n+1]]
    return data.iloc[sim_indices]

# Example: Get top 5 similar restaurants for the first restaurant
similar_restaurants = get_similar_restaurants(0, top_n=5)
similar_restaurants
```
